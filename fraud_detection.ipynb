```python
# the data is highly unbalanced, therefore we are seperating both the classes
fraud = df[df['Class']==1]
norm = df[df['Class']==0]
```

For training our data we will try to balace the data and take equal values for both normal and fraud data

```python
df.groupby('Class').mean()
```

```python
normal = norm.sample(n=492)
normal.head()
# normal data is divided randomly having 492 samples
```

```python
new_data = pd.concat([normal, fraud], axis=0)
new_data.shape
```

```python
# comparing the value of now data to the old one and both are approximately same
new_data.groupby('Class').mean()
```

```python
# now splitting the data into target and features
X = new_data.drop(columns='Class')
y = new_data['Class']
```

```python
X.head()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size=0.2,stratify=y ,random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(y.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

# Checking Accuracy

```python
from sklearn.metrics import accuracy_score
```

```python
X_train_pred = lr.predict(X_train)
training_data_accuracy = accuracy_score(X_train_pred,y_train)
print('accuracy of training data is : ',training_data_accuracy)
```

```python
X_test_pred = lr.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_pred,y_test)
print('accuracy of testing data is : ',testing_data_accuracy)
```

```python
X_test
```

Applying Decision Tree

```python
from sklearn import tree
model = tree.DecisionTreeClassifier(criterion="entropy")
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
# here the accuracy is approx 86%
```
